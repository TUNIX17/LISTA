```python
# prompt: concatenar las columnas arrival_year	arrival_month	arrival_date  y crear una nueva columna llamada Date, como es una fecha  separarlas con un "-"

df['Date'] = df['arrival_year'].astype(str) + "-" + df['arrival_month'].astype(str) + "-" + df['arrival_date'].astype(str)

df
```

```python
# prompt: transformar a fecha la columna Date y mostrar resultados

from datetime import datetime

df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

df
```

```python
# prompt: eliminar columnas arrival_year	arrival_month arrival_date

df = df.drop(columns=['arrival_year', 'arrival_month', 'arrival_date'])
df
```

```python
# prompt: show table

df
```

```python
# prompt: create a new colum called Date_departure = Date + total_of_days

df['Date_departure'] = df['Date'] + pd.to_timedelta(df['total_of_days'], unit='d')

df
```

```python
# prompt: create 2 new colums No_of_week_days and No_of_weekend_days , Date and Date_departure for calculations show results

df['No_of_week_days'] = df.apply(lambda row: (row['Date_departure'] - row['Date']).days - 2*(row['Date_departure'] - row['Date']).days//7, axis=1)
df['No_of_weekend_days'] = df.apply(lambda row: 2*(row['Date_departure'] - row['Date']).days//7, axis=1)

df
```

```python
# prompt: create a colum called DAY_NAME , using the colum Date extract the day of the week (monday, tuesday, etc)

df['DAY_NAME'] = df['Date'].dt.strftime('%A')
df
```

```python
# prompt: create a new colum called month_of_the_year extracted from Date

df['month_of_the_year'] = df['Date'].dt.strftime('%B')
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score, confusion_matrix, classification_report
import shap
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
df = pd.read_csv('Dataset-Hotel+Booking+Cancellation+Prediction.csv')

# Drop irrelevant columns
X = df.drop(columns=['Booking_ID', 'booking_status'])
y = df['booking_status']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Check if index 5839 exists in the testing data
if 5839 < len(X_test):
    # Handle row 5839
    test_row = X_test.iloc[5839]
    test_label = y_test.iloc[5839]

    # Append the row to the training data
    X_train = X_train.append(test_row)
    y_train = y_train.append(pd.Series(test_label))

    # Drop the row from the testing data
    X_test = X_test.drop(index=5839)
    y_test = y_test.drop(index=5839)
else:
    print("Index 5839 is out of bounds for testing data.")

# Train the LabelEncoder for categorical columns
le = LabelEncoder()
le.fit(pd.concat([X_train['type_of_meal_plan'], X_test['type_of_meal_plan']]))

# Transform the categorical columns
X_train['type_of_meal_plan'] = le.transform(X_train['type_of_meal_plan'])
X_test['type_of_meal_plan'] = le.transform(X_test['type_of_meal_plan'])

# Train the OneHotEncoder for remaining categorical columns
categorical_columns = ['market_segment_type', 'room_type_reserved']
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)
ohe.fit(X_train[categorical_columns])

# Transform the remaining categorical columns
X_train_encoded = pd.concat([X_train.drop(categorical_columns, axis=1),
                             pd.DataFrame(ohe.transform(X_train[categorical_columns]),
                                          columns=ohe.get_feature_names_out(),
                                          index=X_train.index)], axis=1)
X_test_encoded = pd.concat([X_test.drop(categorical_columns, axis=1),
                            pd.DataFrame(ohe.transform(X_test[categorical_columns]),
                                         columns=ohe.get_feature_names_out(),
                                         index=X_test.index)], axis=1)

# Train the Random Forest classifier
rf = RandomForestClassifier(n_estimators=100, max_depth=50, min_samples_leaf=1, criterion='gini', random_state=42)
rf.fit(X_train_encoded, y_train)

# Evaluate the model on the testing set
y_pred = rf.predict(X_test_encoded)

# Calculate recall for the positive class ('Canceled')
recall = recall_score(y_test, y_pred, pos_label='Canceled')
print("Recall score:", recall)

# Print the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", cm)

# Print the classification report
print("Classification report:\n", classification_report(y_test, y_pred, target_names=['Not_Canceled', 'Canceled']))

# Calculate feature importance
feature_importances = pd.Series(rf.feature_importances_, index=X_train_encoded.columns)
feature_importances.sort_values(ascending=False).plot(kind='bar')
plt.show()

# Calculate Shapley index for lead_time and avg_price_per_room
explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(X_test_encoded)
shap.summary_plot(shap_values[:, 0], X_test_encoded['lead_time'])
shap.summary_plot(shap_values[:, 1], X_test_encoded['avg_price_per_room'])
```
